# Display the Training Curve

Load the data

```python
import json
import numpy as np
with open('trainer_state.json') as f:
  data = json.load(f)
  logs = data['log_history']
  train_loss = []
  train_steps = []
  eval_loss = []
  eval_steps = []
  for log in logs:
    try:
      train_loss.append(log['loss'])
      train_steps.append(log['step'])
    except:
      try:
        eval_loss.append(log['eval_loss'])
        eval_steps.append(log['step'])
      except:
        print("Final train loss: ", log['train_loss'])
```

Plot the train and validation loss

```python
import matplotlib.pyplot as plt
plt.plot(train_steps, train_loss, label="train")
plt.plot(eval_steps, eval_loss, label="validation")
plt.title("Model loss")
plt.xlabel("Steps")
plt.ylabel("Loss")
plt.legend()
fig = plt.gcf()
plt.show()
```

Save the plot as png file

```python
fig.savefig(f"../../../../figures/training_curve.png", dpi=200, bbox_inches='tight')
```
