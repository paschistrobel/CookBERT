# Extract Vocabulary to Add

Extract all words that occure at least *MIN_OCCURENCES* of times in the whole recipeNLG dataset and add them to the BERT vocab (if not included yet).

```python
MIN_OCCURENCES = 1000

def vocab_occuring_at_least_n_times(n, vocabulary):
  vocab_list = [(k, v) for k, v in vocabulary.items() if v >= n]
  return set(dict(vocab_list))

import json
with open("domain_analysis/recipeNLG_vocabulary.json") as f: # load recipeNLG_vocab
  recipeNLG_vocab = vocab_occuring_at_least_n_times(MIN_OCCURENCES, json.load(f))

with open("CookBERT/further_pretraining/vocabulary/bert-base-uncased-vocab.txt") as f: # load BERT base uncased vocab
  bert_vocab = f.read()
  bert_vocab = bert_vocab.split()

print(f"Number of vocabs in recipeNLG that occure at least {MIN_OCCURENCES} times: {len(recipeNLG_vocab)}")
print(f"Number of vocabs in bert-base-uncased-vocab file: {len(bert_vocab)}")

# create list with vocab that is not included in BERT and thus should be added
already_existing = []
non_existing = []
for word in recipeNLG_vocab:
  if word in bert_vocab:
    already_existing.append(word)
  else:
    non_existing.append(word)

print(f"Number of vocabs that are already in bert-vocab: {len(already_existing)}")
print(f"Number of vocabs that are not in bert-vocab: {len(non_existing)}")
print(non_existing)
```

Save the new vocabulary that should be added to BERT vocab to a txt file:

```python
with open("CookBERT/further_pretraining/vocabulary/vocab_to_add.txt", "w") as f:
  for word in non_existing:
    f.write(word + "\n")
```
